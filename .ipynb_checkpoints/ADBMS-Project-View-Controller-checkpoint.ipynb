```python
choose_date="""
import React, { useState } from 'react';
import axios from 'axios';

const ChooseDate = () => {
    const [selectedDate, setSelectedDate] = useState('');
    const [movies, setMovies] = useState([]);
    const [errorMessage, setErrorMessage] = useState('');

    const handleSubmit = async (event) => {
        event.preventDefault();
        try {
            const response = await axios.post('/', { selected_date: selectedDate });
            setMovies(response.data);
            setErrorMessage('');
        } catch (error) {
            setErrorMessage('Error fetching data');
            console.error('Error:', error);
        }
    };

    return (
        <div>
            <h2>Choose a Date</h2>
            <form onSubmit={handleSubmit}>
                <label htmlFor="selected_date">Select a Date:</label>
                <input
                    type="date"
                    id="selected_date"
                    value={selectedDate}
                    onChange={(event) => setSelectedDate(event.target.value)}
                    required
                />
                <br /><br />
                <input type="submit" value="Submit" />
            </form>

            {errorMessage && <p>{errorMessage}</p>}

            <div id="movies_list">
                <h3>Movies Available on {selectedDate}</h3>
                <ul>
                    {movies.length > 0 ? (
                        movies.map(movie => (
                            <li key={movie.id}>
                                <strong>{movie.name}</strong> - {movie.lead_actor} (Duration: {movie.duration} minutes)
                            </li>
                        ))
                    ) : (
                        <li>No movies available on {selectedDate}</li>
                    )}
                </ul>
            </div>
        </div>
    );
};

export default ChooseDate;
"""
```

```python



from flask import Flask, request, jsonify
import mysql.connector

app = Flask(__name__)

# HTML string for the choose_date page


@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        selected_date = request.form['selected_date']
        movies, theaters = get_movies_and_theatres(selected_date)
        return render_template('movies_and_theatres.html', movies=movies, theatre=theatres)
    return choose_date_slots_html


@app.route('/schedule', methods=['GET'])
def get_schedule():
    try:
        
        print("inside the schedule part")
        # Connect to the database
        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()

        # Extract the date from the query parameter
        date = request.args.get('date')
        
        print("Date is ",date)
        query = """
            SELECT movie.name AS movie, theatre.name AS theater
            FROM running
            INNER JOIN movie ON running.movie_id = movie.id
            INNER JOIN theatre ON running.theatre_id = theatre.id
            WHERE %s 
        """

        # Query to fetch movies and theaters for the given date
        query = """
            SELECT movie.name AS movie, theatre.name, theatre.id AS theater
            FROM running
            INNER JOIN movie ON running.movie_id = movie.id
            INNER JOIN theatre ON running.theatre_id = theatre.id
            WHERE %s BETWEEN running.start_date AND running.end_date
        """
        
        
        cursor.execute(query, (date,))
        
        
        print("query executed")
        
        schedule = cursor.fetchall()

        # Close the cursor and connection
        cursor.close()
        connection.close()

        # Convert the result to a list of dictionaries
        schedule_list = [{'movie': row[0], 'theater': row[1], 'theatre_id': row[2]} for row in schedule]
        
        print("Schedule list is ")
        print(schedule_list)
        
        print("")

        return jsonify(schedule_list)

    except Exception as e:
        return jsonify({'error': str(e)}), 500
    

@app.route('/get-slots', methods=['GET'])
def get_show_slots():
    try:
        print("we are here at least")
        # Get movie and date from request parameters
        theatre = request.args.get('theatre_id')
        show_date = request.args.get('date')

        
        print("theatre is ",theatre)
        print("show date is ",show_date)
        
        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()
        
        print("Connection established")

        
        query = """
            SELECT show_number, theatre_id, time(show_time_start), time(show_time_end)
            FROM show_timing_seats_details
        """
        quer = """
            SELECT show_number, show_time_start, show_time_end
            FROM show_timing_seats_details
            WHERE theatre_id = %s AND show_date = '%s'
        """
        
        print("Hi there")
        query = "SELECT show_number, theatre_id, time(show_time_start), time(show_time_end) FROM show_timing_seats_details WHERE theatre_id = {} AND show_date = '{}' ".format(theatre,show_date)
        
        print("query modified")
        
        print("Query is",query)
        cursor.execute(query)
        
        print("query executed")
#        cursor.execute(quer, (theatre, show_date,))
        slots_data = cursor.fetchall()

        cursor.close()
        connection.close()

        slots_json = []
        for slot in slots_data:
            slots_json.append({
                'show_number': slot[0],
                'theatre_id': slot[1],
                'start_time': str(slot[2]),  # Convert time to string
                'end_time': str(slot[3])      # Convert time to string
            })
            
        print(slots_json)

        return jsonify(slots_json)


    except Exception as e:
        return jsonify({'error blah blah': str(e)}), 500
    
    
@app.route('/make-booking', methods=['PUT'])
def make_booking():
    try:
        print("we are here at least")
        # Get movie and date from request parameters
        theatre_id = request.args.get('theatre_id')
        show_date = request.args.get('date')
        show_number= request.args.get('show_number')
        name=request.args.get('name')
        
        print("param values are ",theatre_id,show_date,show_number,name)

        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()

        query = "INSERTINTO booking(show_theatre_id,show_date,show_number,customer_name)VALUES({},'{}',{},'{}')".format(theatre_id,show_date,show_number,name)

        print("Query is ",query)
        
        cursor.execute(query)
        
        print("Query executed")
        
#        cursor.execute(query, (theatre_id, show_date))
#        slots_data = cursor.fetchall()

        cursor.close()
        connection.close()


            
#        print(slots_json)

        return jsonify({'Booking_Status':'Booking Successful'})


    except Exception as e:
        return jsonify({'Booking_Status': 'Booking Not Successful'})
#        return jsonify({'Booking_Status': 'Booking Not Successful'}), 500


if __name__ == '__main__':
    app.run(debug=True,use_reloader=False)
```
